````python
import json

with open('./datas/전국박물관미술관정보표준데이터.json', 'r', encoding='utf-8') as f:
    json_data = json.load(f)
````

        - json_data['records']: List. 각 variable은 하나의 Row(행)을 의미하며, {Column(열)명: data}의 형식으로 구성 되어있습니다.

````python
import pandas as pd

fields = json_data['fields']
columns = [field['id'] for field in fields]

# json_data에서 레코드 정보 추출
records = json_data['records']
data = []

# 각 레코드에서 필드에 해당하는 데이터 추출
for record in records:
    row_data = [record[column] for column in columns]
    data.append(row_data)

df_target = pd.DataFrame(data, columns=columns)
df_target
````
