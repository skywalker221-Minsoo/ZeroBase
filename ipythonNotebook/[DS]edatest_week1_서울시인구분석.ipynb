### **문제 3-4)** 자치구별 외국인비율-세대당인구를 Scatter plot에 나타내고, 상관관계에 따른 Regression Line을 시각화 하세요.    (10점)

- 예시
<br></br>
    <img src="./[문제보기용 이미지]images/01-3-4-001.png" style="margin:auto" width=400/>

````python
import numpy as np

x = df_target_barh["외국인비율"]
y = df_target_barh['세대당인구']

fp1 = np.polyfit(x, y, 1)
f1 = np.poly1d(fp1)
fx = np.linspace(0, 9, 100)

plt.scatter(x, y)
plt.plot(fx, f1(fx), ls="dashed", lw=3, color="g")
plt.xlabel('외국인비율')
plt.ylabel('세대당인구')
plt.grid(True)
plt.show()
````
