<br></br>
    - hint2: 내용이 일치하는 데이터를 삭제하기 위한 방법은 drop_duplicate, groupby, pivot_table 등 다양한 방법이 있습니다.
<br></br>
    - hint3: Medal Index의 정렬 순서를 'Gold'-'Silver'-'Bronze'로 하기 위해서 sort_index의 'key' argument에 dictionary를 활용할 수 있습니다.
        - 이 외에 sort_index의 다른 arguments도 확인해 보세요
<br></br>
<br></br>
    - 참고1: 예시와 같이 컬럼명은 City 등 아무 컬럼명이든 상관 없습니다. 하나의 컬럼만 보여주세요.
<br></br>
    - 참고2: 추후 문제풀이시 문제 없게 하기 위해서 현재데이터를 deep copy하여 진행하시는 것을 추천드립니다. 
        - 하지 않아도 문제는 없으나 코드 작성 방법에 따라 다음 문제 풀이시 다시 load하고 전처리 진행해야 할 수도 있습니다.

````python
df_KOR = df_Olympic.copy()
df_KOR = df_KOR[df_KOR['Country'] == 'Korea, South']
df_KOR = df_KOR.drop_duplicates(subset=['City', 'Year', 'Sport', 'Discipline', 'Event', 'Event_gender', 'Medal'])

df_KOR = df_KOR.groupby(['Year', 'Medal']).size().reset_index(name="Count")

df_KOR = df_KOR.rename(index={0: 2, 1: 0, 2: 1})
df_KOR = df_KOR.rename(index={3: 5, 4: 3, 5: 4})
df_KOR = df_KOR.rename(index={6: 8, 7: 6, 8: 7})
df_KOR = df_KOR.rename(index={9: 11, 10: 9, 11: 10})
df_KOR = df_KOR.rename(index={12: 14, 13: 12, 14: 13})
df_KOR = df_KOR.rename(index={15: 17, 16: 15, 17: 16})
df_KOR = df_KOR.rename(index={18: 20, 19: 18, 20: 19})
df_KOR = df_KOR.rename(index={21: 23, 22: 21, 23: 22})
df_KOR = df_KOR.sort_index(ascending=True)

df_KOR = df_KOR.set_index(['Year', 'Medal'])

df_KOR
````
