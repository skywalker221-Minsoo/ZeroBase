문제 6. 

위도, 경도 정보를 이용하여 미왕빌딩에서 1킬로 이내에 위치한 주유소 정보를 검색하세요. 

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 미왕빌딩으로 부터의 거리 (km)

제출 6.
- 쿼리, 결과

```python
gmaps_key = "AIzaSyC8awub1OYEUlUzUboX4H0vq2qqOLEtXvw"
gmaps = googlemaps.Client(key=gmaps_key)

# 미왕빌딩의 주소 정보를 활용하여 위도, 경도 정보를 알아냄
address = "서울특별시 강남구 테헤란로 518"
geo = gmaps.geocode(address, language="ko")[0]
lat = geo['geometry']['location']['lat']
lng = geo['geometry']['location']['lng']

from math import radians, cos, sin, asin, sqrt

# haversine: 지구를 구(sphere)로 가정하여 두 지점의 좌표(경도, 위도)를 이용하여 지구 표면을 따라 측정한 거리
# 대원 거리(great-circle distance)를 구하는데 활용
def haversine(lat1, lon1, lat2, lon2):
    # Convert degrees to radians
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # Haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a))
    # 거리 = 원의 반지름 * 떨어진 두 점의 각도
    km = 6371 * c
    return km

oil_df['lat'] = oil_df['lat'].replace('', np.nan).astype('float')
oil_df['lng'] = oil_df['lng'].replace('', np.nan).astype('float')

# 주유소 데이터프레임(df)에서 위도, 경도 정보 추출
latitudes = oil_df['lat'].tolist()
longitudes = oil_df['lng'].tolist()

# 모든 주유소와 미왕빌딩 간의 거리 계산
distances = [haversine(lat, lng, lat1, lng1) for lat1, lng1 in zip(latitudes, longitudes)]

step6 = oil_df.copy()
# step6 = step6.reset_index(drop=True)
step6['id'] = step6.index+1
step6['거리'] = distances

step6 = step6[step6['거리'] <= 1]
step6['브랜드'] = step6['브랜드'].apply(lambda x: 'SK에너지' if x == 1 else '현대오일뱅크' if x==2 else 'GS칼텍스' if x==3
                                                else 'S-OIL' if x==4 else '알뜰주유소' if x==5 else '자가상표')
step6[['id', '브랜드', '주유소명', '주소', '거리']]
```
