````python
import statsmodels.api as sm

def calculate_regression_coefficients(series):
    X = np.arange(len(series))  # '기간'을 독립 변수로 설정
    X = sm.add_constant(X)  # 상수항을 추가
    y = series.values  # 종속 변수

    model = sm.OLS(y, X)  # 선형 회귀모델 생성
    results = model.fit()  # 회귀모델 피팅
    return results.params[1]  # 기울기(회귀계수) 반환

df_result = df_target.sort_values(by=['구분1', '구분2']).reset_index(drop=True)

df_result['전체'] = df_result.groupby('구분2')['전체'].transform(calculate_regression_coefficients)
df_result['남자'] = df_result.groupby('구분2')['남자'].transform(calculate_regression_coefficients)
df_result['여자'] = df_result.groupby('구분2')['여자'].transform(calculate_regression_coefficients)

df_result = round(df_result.groupby(['구분1', '구분2']).mean(), 6)

df_result = df_result.drop('기간', axis=1)
df_result = df_result.reset_index()
df_result
````
